## 3.2.1. Applicability of tests

Different statistical tests have different assumptions that need to be met to ensure the validity and reliability of the results.


**T- test:**
The t-test is used to compare the means of two groups or independent samples. In this case, the t-test was used to compare nurses' salaries in different sectors.

- [x] Independence: The observations are independent of each other.
- [ ] Normality: The distribution of the data within each group should be approximately normal. For large sample sizes (n > 30), the t-test is robust to violations of normality.
- [ ] Homogeneity of Variance (Equal Variance): The variance of the data in the different groups should be roughly equal. Violation of this assumption may affect the results.

**Anova:**
The Anova is uses to compare the means of three groups or more groups. The test compares the means of the salary distributions, assuming the data follows a normal distribution and the variances are approximately equal.
In this project the Anova was used to compare the the salaries of nurse in different sectors or with different levels of education. Before using the one-way ANOVA test, it is crucial to check if the assumptions of the test are met. The main assumptions of the one-way ANOVA include:
- [x] Independence: The observations are independent of each other.
- [ ] Normality: The data within each group should be approximately normally distributed. For large sample sizes, ANOVA is robust to violations of normality.
- [ ] Homogeneity of Variance (Equal Variance): The variance of the data in the different groups should be approximately equal.

If the data does not meet these assumptions, it may be more appropriate to use non-parametric tests like the Kruskal-Wallis test

**Chi-Squared:**
The Chi-Squared is used to association or independence between of  two categorical variable.
- [x] Independence: The observations should be independent of each other.
- [ ] Expected Cell Frequency: The expected frequency count for each cell should be at least 5. Violation of this assumption may affect the validity of the test.


**Correlation:**

- [ ] Linearity: The relationship between the two continuous variables should be approximately linear.
- [x] Independence: The observations should be independent of each other.
- [ ] Normality: For Pearson correlation, the variables should be approximately normally distributed. For Spearman or Kendall rank correlation, this assumption is less strict, but the data should not have extreme outliers.


Regression:

- [ ] Linearity: The relationship between the dependent variable and each independent variable should be linear.
- [x] Independence: The observations should be independent of each other.
- [ ] Homoscedasticity (Constant Variance): The residuals (the differences between the predicted and observed values) should have constant variance across all levels of the independent variables.
- [ ] Normality of Residuals: The residuals should be approximately normally distributed.
- [ ] No Multicollinearity: The independent variables should not be highly correlated with each other.


As we see, the assumptions needed to apply any of these tests are not met. This is unfortunate. What this means is that if we apply these tests, we cannot trust in the results, because the distribution of our data has no homogeneus variance (Heteroscedasticity). # 3.3. Non Parametric Tests

non-parametric tests are valuable in situations where the assumptions of parametric tests (e.g., t-tests, ANOVA, linear regression) are violated, and they provide a robust and reliable way to analyze data without assuming a specific distribution. However, non-parametric tests might have less statistical power than their parametric counterparts


### **Wilkoxon signed-rank test:**
wilkinson is used to compare the medians of two different groups when there are something in the data that is not normally distributed.
- [ ] Dependent Samples: The data should consist of paired observations, where each individual or subject has two related measurements.
- [x]  Similarity of Distributions: The differences between the paired observations should have a similar shape for both positive and negative values. In other words, the distribution of the differences should not be heavily skewed in one direction.
- [ ]  Symmetry: The underlying distributions of the differences should be symmetric. This assumption is related to the distribution of ranks in the test.
- [x]  Independence: The paired observations should be independent of each other.
- [x]  Ordinal Data: While the Wilcoxon test can be used with ordinal or continuous data, it is more powerful and appropriate for continuous data. For ordinal data, you may consider other non-parametric tests like the Sign test or the McNemar test, depending on the nature of the data.
- [ ]  Random Sampling: The paired observations should be a random sample from the population of interest.

**This test is used when compared between two groups.**

### Kruskal-Wallis test
we can use the Kruskal-Wallis test. The Kruskal-Wallis test is a non-parametric test used to compare the median ranks of three or more independent groups or samples. In this case, we want to compare the salaries of nurses across different countries, which are independent groups.

The assumptions that have to be met are:

- [x] Independence: The observations in each group should be independent of each other. This means that the data points in one group should not be related to or dependent on the data points in other groups.

- [ ] Random Sampling: The data should be obtained through a random sampling process from each group. The samples should be representative of the populations they are drawn from.

- [x] Ordinal Data: The Kruskal-Wallis test is most appropriate when the data is at least ordinal, meaning the values can be ranked. It can also be used with continuous data, but it is less powerful than ANOVA if the data are truly continuous and normally distributed.

- [x] Similar Shape of Distributions: The distributions of the data in each group should have similar shapes. This means that the spread of data and skewness should be roughly the same across all groups.

- [ ] Homogeneity of Variance: The variance of the data in each group should be roughly equal. While the Kruskal-Wallis test is less sensitive to unequal variances compared to ANOVA, extreme differences in variances might affect the test's performance.

**This test is used when comparing between multiple groups**

## Salaried vs Self-Employed Salaries

To test this hypothesis, we perform the Wilcoxon rank-sum test for each pair of countries. The loop compares the salary distributions of nurses between each pair of countries using the ranksums function from scipy.stats. The test returns a p-value, which indicates the probability of obtaining the observed results if the null hypothesis (no significant difference) is true.

If any of the calculated p-values are less than the chosen significance level (alpha, typically set to 0.05), we reject the null hypothesis and conclude that there is a significant difference in the average salary of nurses between at least one pair of countries. Otherwise, if all p-values are greater than alpha, we fail to reject the null hypothesis, suggesting that there is no significant difference in nurse salaries across countries.

````python
from scipy.stats import ranksums

# Assuming your data is loaded into the DataFrame called 'df'

# Subset the data to include only nurses
nurses_df = df[df['Employment'] == 'Salaried']

# Perform the Wilcoxon rank-sum test for each pair of countries
countries = nurses_df['Country'].unique()
num_countries = len(countries)
p_values = []

for i in range(num_countries - 1):
    for j in range(i + 1, num_countries):
        country1 = countries[i]
        country2 = countries[j]
        salary_country1 = nurses_df[nurses_df['Country'] == country1]['Income at 2015 GDP price level, in US$']
        salary_country2 = nurses_df[nurses_df['Country'] == country2]['Income at 2015 GDP price level, in US$']
        result = ranksums(salary_country1, salary_country2)
        p_values.append((country1, country2, result.pvalue))

# Set the significance level (alpha)
alpha = 0.05

# Check if any p-value is less than alpha
significant_difference = any(p_value < alpha for _, _, p_value in p_values)

if significant_difference:
    print("There is a significant difference in the average salary of nurses between different Types of Employment: salaried vs self-employed.")
else:
    print("There is no significant difference in the average salary of nurses between different Types of Employment: salaried vs self-employed.")
````

## Salaries of nurses in different countries

To test this hypothesis, we first subset the data to include only nurses. Then, we group the data by the 'Country' column and extract the salaries of nurses for each country. We use the Kruskal-Wallis test to compare the salary distributions across these groups. The test returns a p-value, which indicates the probability of obtaining the observed results if the null hypothesis (no significant difference) is true.

If the p-value is less than the chosen significance level (alpha, typically set to 0.05), we reject the null hypothesis and conclude that there is a significant difference in the average salary of nurses between different countries. On the other hand, if the p-value is greater than alpha, we fail to reject the null hypothesis, suggesting that there is no significant difference in nurse salaries across countries.

````python
from scipy.stats import kruskal

# Assuming your data is loaded into the DataFrame called 'df'

# Subset the data to include only nurses
nurses_df = df[df['Professional - Type of Employment'] == 'hospital nurses - Salaried']

# Perform the Kruskal-Wallis test
result = kruskal(*[group['Income at 2015 GDP price level, in US$'] for country, group in nurses_df.groupby('Country')])

# Extract the p-value from the result
p_value = result.pvalue

# Set the significance level (alpha)
alpha = 0.05

# Compare the p-value to the significance level
if p_value < alpha:
    print("There is a significant difference in the average salary of nurses between different countries.")
else:
    print("There is no significant difference in the average salary of nurses between different countries.")
````

## Salaries over time

To test this hypothesis, we first subset the data to include only nurses. Then, we group the data by the 'Year' column and extract the salaries of nurses for each year. We use the Kruskal-Wallis test to compare the salary distributions across these groups (i.e., different years). The test returns a p-value, which indicates the probability of obtaining the observed results if the null hypothesis (no significant variation) is true.

If the p-value is less than the chosen significance level (alpha, typically set to 0.05), we reject the null hypothesis and conclude that there is a significant variation in the average salary of nurses over time. If the p-value is greater than alpha, we fail to reject the null hypothesis, suggesting that there is no significant variation in the average salary of nurses over time.

````python
from scipy.stats import kruskal

# Assuming your data is loaded into the DataFrame called 'df'

# Subset the data to include only nurses
nurses_df = df[df['Variable'] == 'Remuneration of hospital nurses']

# Perform the Kruskal-Wallis test
result = kruskal(*[group['Income at 2015 GDP price level, in US$'] for year, group in nurses_df.groupby('Year')])

# Extract the p-value from the result
p_value = result.pvalue

# Set the significance level (alpha)
alpha = 0.05

# Compare the p-value to the significance level
if p_value < alpha:
    print("There is a significant variation in the average salary of nurses over time.")
else:
    print("There is no significant variation in the average salary of nurses over time.")
````

## Nurse salaries vs average salaries of the country

We can use a T-Test, having in mind that in the subset of Nurse salaries, if considering only the hospital nurses, the data has not so much variation in variance.


````python
from scipy.stats import ttest_1samp

# Assuming your data is loaded into the DataFrame called 'df'

# Subset the data to include only nurses
nurses_df = df[df['Professional - Type of Employment'] == 'hospital nurses - Salaried']

# Extract the nurse salaries
nurse_salaries = nurses_df['Income per average wage']

# Perform the one-sample t-test
population_mean = df['Income per average wage'].mean()
result = ttest_1samp(nurse_salaries, population_mean)

# Extract the p-value from the result
p_value = result.pvalue

# Set the significance level (alpha)
alpha = 0.05

# Compare the p-value to the significance level
if p_value < alpha:
    print("Nurse salaries differ significantly from the average salaries of the country.")
else:
    print("Nurse salaries do not differ significantly from the average salaries of the country.")
````
